pd.to_numeric(..., errors='coerce'): đảm bảo mọi giá trị không chuyển được thành số sẽ bị biến thành NaN, giúp tránh lỗi.

np.where((b == 0) | (pd.isna(b)), np.nan, a / b): xử lý chia cho 0 và chia cho NaN → trả về NaN, tránh lỗi chia sai.

```python
import numpy as np
import pandas as pd

def safe_div(a, b):
    a = pd.to_numeric(a, errors="coerce")
    b = pd.to_numeric(b, errors="coerce")
    return np.where((b == 0) | (pd.isna(b)), np.nan, a / b)
```

```python
# Hàm tìm cột chứa từ khóa
def find_col(keyword):
    matches = [col for col in df.columns if keyword.lower() in col.lower()]
    if matches:
        return matches[0]
    else:
        print(f"Không tìm thấy cột chứa: {keyword}")
        return None

# Bảng cân đối kế toán
current_assets       = find_col("CĐKT. TÀI SẢN NGẮN HẠN")
current_liabilities  = find_col("CĐKT. Nợ ngắn hạn")
cash                 = find_col("CĐKT. Tiền và tương đương tiền")
short_term_investments = find_col("CĐKT. Đầu tư tài chính ngắn hạn")
accounts_receivable  = find_col("CĐKT. Các khoản phải thu ngắn hạn")
inventory            = find_col("CĐKT. Hàng tồn kho, ròng")

total_assets         = find_col("CĐKT. TỔNG CỘNG TÀI SẢN")
fixed_assets         = find_col("CĐKT. Tài sản cố định")
total_liabilities    = find_col("CĐKT. NỢ PHẢI TRẢ")
equity               = find_col("CĐKT. VỐN CHỦ SỞ HỮU")

# KQKD
revenue              = find_col("KQKD. Doanh thu thuần")
net_income           = find_col("KQKD. Lợi nhuận sau thuế thu nhập doanh nghiệp")
income_tax           = find_col("KQKD. Chi phí thuế thu nhập doanh nghiệp")
interest_expense     = find_col("KQKD. Trong đó: Chi phí lãi vay")

# LCTT
depreciation         = find_col("LCTT. Khấu hao TSCĐ")
purchase             = find_col("KQKD. Chi phí tài chính")
```

```python
# Chuẩn hóa dữ liệu + sort theo (Mã, Năm)

import numpy as np
import pandas as pd

def safe_div(a, b):
    a = pd.to_numeric(a, errors="coerce")
    b = pd.to_numeric(b, errors="coerce")
    return np.where((b == 0) | (pd.isna(b)), np.nan, a / b)

# Copy các cột cần dùng
keep_cols = [
    "Mã", "Tên công ty", "Sàn",
    "Ngành ICB - cấp 1", "Ngành ICB - cấp 2", "Ngành ICB - cấp 3", "Ngành ICB - cấp 4",
    "Năm",
    current_assets, current_liabilities, cash, short_term_investments, accounts_receivable, inventory,
    total_assets, fixed_assets, total_liabilities, equity,
    revenue, net_income, income_tax, interest_expense,
    depreciation, purchase
]

df_base = df[keep_cols].copy()

# Ép kiểu số cho các cột numeric
id_cols = ["Mã","Tên công ty","Sàn","Ngành ICB - cấp 1","Ngành ICB - cấp 2","Ngành ICB - cấp 3","Ngành ICB - cấp 4"]
for c in df_base.columns:
    if c not in id_cols:
        df_base[c] = pd.to_numeric(df_base[c], errors="coerce")

df_base = df_base.sort_values(["Mã","Năm"]).reset_index(drop=True)
print("df_base shape:", df_base.shape)
```

```python
# Tính EBITDA và COGS proxy

# EBITDA = LNST + Thuế TNDN + Lãi vay + Khấu hao
df_base["EBITDA"] = (
    df_base[net_income]
    + df_base[income_tax]
    + df_base[interest_expense]
    + df_base[depreciation]
)

# Tồn kho đầu kỳ = tồn kho năm trước (theo Mã)
df_base["INV_lag"] = df_base.groupby("Mã")[inventory].shift(1)

# COGS proxy = Tồn kho đầu kỳ + Chi phí mua hàng - Tồn kho cuối kỳ
df_base["COGS_proxy"] = (
    df_base["INV_lag"]
    + df_base[purchase]
    - df_base[inventory]
)
```

```python
df_ratios = pd.DataFrame({
    "Mã": df_base["Mã"],
    "Tên công ty": df_base["Tên công ty"],
    "Sàn": df_base["Sàn"],
    "Ngành ICB - cấp 1": df_base["Ngành ICB - cấp 1"],
    "Ngành ICB - cấp 2": df_base["Ngành ICB - cấp 2"],
    "Ngành ICB - cấp 3": df_base["Ngành ICB - cấp 3"],
    "Ngành ICB - cấp 4": df_base["Ngành ICB - cấp 4"],
    "Năm": df_base["Năm"],

    # I. Thanh khoản
    "Current_Ratio": safe_div(df_base[current_assets], df_base[current_liabilities]),
    "Quick_Ratio": safe_div(df_base[cash] + df_base[short_term_investments] + df_base[accounts_receivable],
                            df_base[current_liabilities]),
    "Cash_Ratio": safe_div(df_base[cash], df_base[current_liabilities]),

    # II. Đòn bẩy
    "Debt_Equity": safe_div(df_base[total_liabilities], df_base[equity]),
    "Net_Leverage": safe_div(df_base[total_liabilities] - df_base[cash], df_base["EBITDA"]),

    # III. Hiệu quả
    "Asset_Turnover": safe_div(df_base[revenue], df_base[total_assets]),
    "Fixed_Asset_Turnover": safe_div(df_base[revenue], df_base[fixed_assets]),
    "Inventory_Turnover": safe_div(df_base["COGS_proxy"], df_base[inventory]),
    "Receivables_Turnover": safe_div(df_base[revenue], df_base[accounts_receivable]),

    # IV. Quy mô
    "ln_Total_Assets": np.log(df_base[total_assets] + 1),
    "ln_Revenue": np.log(df_base[revenue] + 1),

    # Biến đánh giá (không dùng để phân cụm)
    "ROA": safe_div(df_base[net_income], df_base[total_assets]),
    "ROE": safe_div(df_base[net_income], df_base[equity]),
    "ROS": safe_div(df_base[net_income], df_base[revenue]),
})

# loại trùng theo Mã-Năm (phòng trường hợp dữ liệu lặp)
df_ratios = df_ratios.drop_duplicates(subset=["Mã","Năm"]).reset_index(drop=True)

df_ratios.head(), df_ratios.shape
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import os
import numpy as np

# Load dữ liệu đầu vào
file_path = "../data/processed/financial_ratios_2020_2024.csv"
df = pd.read_csv(file_path)

# Các nhóm chỉ số để chuẩn hóa
groups = {
    "Liquidity_Score": ["Current_Ratio", "Quick_Ratio", "Cash_Ratio"],
    "Leverage_Score": ["Debt_Equity", "Net_Leverage"],
    "Efficiency_Score": ["Asset_Turnover", "Fixed_Asset_Turnover", "Inventory_Turnover", "Receivables_Turnover"],
    "Size_Score": ["ln_Total_Assets", "ln_Revenue"],
    "Profitability_Score": ["ROA", "ROE", "ROS"]
}

# Copy để tránh ảnh hưởng dữ liệu gốc
df_scores = df.copy()

# === TÍNH Z-SCORE CỤC BỘ THEO (Năm, Ngành ICB cấp 1) ===
for new_col, cols in groups.items():

    def compute_local_z(x):
        x = x.astype(float)
        if x.shape[0] < 3:   # quá ít DN → không chuẩn hóa
            return pd.Series([np.nan] * x.shape[0], index=x.index)

        scaler = StandardScaler()
        z = scaler.fit_transform(x)
        return pd.Series(z.mean(axis=1), index=x.index)

    df_scores[new_col] = (
        df_scores
        .groupby(["Năm", "Ngành ICB - cấp 1"])[cols]
        .apply(compute_local_z)
        .reset_index(level=[0,1], drop=True)
    )

# Chỉ giữ các cột cần cho phân cụm & dashboard
cols_to_keep = [
    "Mã", "Tên công ty", "Sàn", "Ngành ICB - cấp 1", "Năm",
    "Liquidity_Score", "Leverage_Score", "Efficiency_Score",
    "Size_Score", "Profitability_Score"
]

df_cluster = df_scores[cols_to_keep].copy()

# Lưu file
out_path = "../data/processed/financial_scores_for_clustering.csv"
os.makedirs(os.path.dirname(out_path), exist_ok=True)
df_cluster.to_csv(out_path, index=False, encoding="utf-8-sig")

print("✅ Đã lưu file chuẩn hóa cục bộ:", out_path)
```

```python
import pandas as pd

# Load lại dữ liệu đã chuẩn hóa
file_path = "../data/processed/financial_scores_for_clustering.csv"
df_cluster = pd.read_csv(file_path)

# Các cột điểm nhóm
score_cols = ["Liquidity_Score", "Leverage_Score", "Efficiency_Score", "Size_Score", "Profitability_Score"]

# Tính tỷ lệ NaN trên mỗi cột theo từng ngành
nan_by_industry = df_cluster.groupby("Ngành ICB - cấp 1")[score_cols].apply(lambda x: x.isna().mean())

# Hiển thị tỷ lệ %
nan_by_industry_percent = (nan_by_industry * 100).round(2)
print("📊 Tỷ lệ % thiếu dữ liệu theo từng ngành:")
display(nan_by_industry_percent)
```
