```python
import pandas as pd
import numpy as np

file_path = "../data/processed/financial_scores_for_clustering.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
def classify_state(z):
    if pd.isna(z):
        return "Không xác định"
    elif z >= 0.5:
        return "Cao"
    elif z <= -0.5:
        return "Thấp"
    else:
        return "Trung bình"
```

```python
score_cols = [
    "Liquidity_Score",
    "Leverage_Score",
    "Efficiency_Score",
    "Size_Score",
    "Profitability_Score"
]

for col in score_cols:
    state_col = col.replace("_Score", "_State")
    df[state_col] = df[col].apply(classify_state)
```

```python
df[
    ["Mã", "Tên công ty", "Sàn", "Ngành ICB - cấp 1", "Năm"]
    + score_cols
    + [c.replace("_Score", "_State") for c in score_cols]
].head(10)
```
